**2.7.** Дана функция:

$$
 \begin{cases}
    x^2 + 2x + 6, & x < 0  \\
    x + 6, & 0 \le x \le 2 \\
    x^2 + 4x - 4, & x \ge 2
 \end{cases}
$$

Постройте массив из её значений на  $-3 \le x \le 3$.

```python
func = lambda x: x**2 + 2*x + 6 if x < 0 \
    else (x + 6 if x >= 0 and x <= 2 \
    else x**2 + 4*x - 4)
```

```python
X = np.arange(-3, 3.1, 0.1)
res = np.array(list(map(func, X)))
res
```

**2.11.** Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
def sample(x, c):
    assert len(x) > 0
    
    s = np.sum(x)
    res = []
    for _ in range(c):
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val:
            cur += x[idx]
            idx += 1
            
        res.append(idx)
    return res

            
sample([50, 3, 1, 7, 20], 5)    
```

Реализуйте данный алгоритм более оптимальным способом, без использования циклов.

```python
def sample2(x, c):
    
    xs = np.cumsum(x).reshape((1, len(x))).repeat(c, axis=0)
    s = np.repeat(np.sum(x), c)
    find_vals = lambda x: x * np.random.random() 
    vals =  np.array(list(map(find_vals, s))).reshape((c, 1))
    
    res = np.sum(xs <= vals, axis=1)
    return res
```

```python
x = [50, 3, 1, 7, 20]
c = 5
sample2(x, c)
```
